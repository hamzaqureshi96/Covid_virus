```python
df1=pd.read_csv(raw_data/'novel-corona-virus-2019-dataset/covid_19_data.csv')
```

```python
df1=df1.drop(columns=['SNo' , 'Last Update'])
```

```python
df1.columns
```

```python
df_grp=df1.groupby(['ObservationDate', 'Province/State', 'Country/Region']).sum()
df_grp.head()
```

```python
df_sort=df1.sort_values(by=['Country/Region' , 'ObservationDate'])
```

```python
df_index=df_sort.set_index('ObservationDate')
df_index.index=pd.to_datetime(df_index.index)
```

```python
def resample(df_data, data_cols, cntry_col, freq):
    df_resample_concat=pd.DataFrame([])
    for cntry in df_data['Country/Region'].unique():
        df=df_data[df_data['Country/Region']==cntry]
        d = {}
        d_data = {i: 'sum' for i in data_cols}
        d_cntry = {i:"first" for i in cntry_col}
        d = {**d_data, **d_cntry}
        df_resample=df.resample(freq).agg(d)
        df_resample_concat=df_resample_concat.append(df_resample)
    return df_resample_concat
```

```python
df_resample=resample(df_index,['Confirmed', 'Deaths', 'Recovered'] , ['Country/Region'], '1D')
```

```python
for cntry in df_resample['Country/Region'].unique():
    df= df_resample[df_resample['Country/Region']==cntry]
    print(cntry, df['Confirmed'].max())
```

## Downloading latest available data

```python
data_2=pathlib.Path(raw_data/'data.humdata/')
```

```python
files=['time_series_covid19_confirmed_global.csv', 'time_series_covid19_deaths_global.csv','time_series_covid19_recovered_global.csv']

for file in files:
    url = f'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv&filename={file}'
    r = requests.get(url, allow_redirects=True)
    open(data_2/file, 'wb').write(r.content)
```

```python
df_conf=pd.read_csv(data_2/files[0])
df_deaths=pd.read_csv(data_2/files[1])
df_recover=pd.read_csv(data_2/files[2])

df_conf.shape, df_deaths.shape, df_recover.shape
```
